Vemos un comportamiento interesante entre la variable VENTO, VELOCIDADE HORARIA que es la variable respuesta y las otras variables donde podemos observar que solo existe una correlación significativa con la variable VENTO, RAJADA MAXIMA. 

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np

df_treino = df_treino.replace([np.inf, -np.inf], np.nan)  # Reemplazar infinitos con NaN

df_treino = df_treino.fillna(0)
df_treino = df_treino.apply(pd.to_numeric, errors='coerce')  # Convertir a numérico

# Crear una lista para almacenar los VIFs
vif_values = []

# Iterar a través de las variables independientes
for i, column in enumerate(df_treino.columns):
    if column != 'VENTO, DIREï¿½ï¿½O HORARIA (gr) (ï¿½ (gr))':  # Excluir la variable 'X1' que estamos calculando
        exog = df_treino[df_treino.columns.difference(['VENTO, DIREï¿½ï¿½O HORARIA (gr) (ï¿½ (gr))', column])]  # Excluir 'X1' y la variable actual
        vif = variance_inflation_factor(exog.values, i)
        vif_values.append((column, vif))
        
# Imprimir los VIFs para todas las variables
for variable, vif in vif_values:
    print(f"VIF para '{variable}': {vif}")
```

## KNN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

```python
X = df_filtered.drop(columns = ['VENTO, VELOCIDADE HORARIA (m/s)'])
y = df_filtered["VENTO, VELOCIDADE HORARIA (m/s)"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

Separamos el conjunto de datos en train y test donde se define el valor de x y la predición para y.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_squared_log_error

rmsle_val = []
best_rmsle = 1.0

for k in range(20):
    k = k + 1
    knn = KNeighborsRegressor(n_neighbors=k)

    knn.fit(X_train, y_train) 
    y_pred = knn.predict(X_test)
    rmsle = np.sqrt(mean_squared_log_error(y_test, y_pred))
    if (rmsle < best_rmsle):
        best_rmsle = rmsle
        best_k = k
    rmsle_val.append(rmsle)
    print('RMSLE value for k=', k, 'is:', rmsle)

print(f"Best RMSLE: {best_rmsle}, Best k: {best_k}")
```

El menor error se produce cuando K = 20, indica un error relativo entre los valores predichos y los actuales del 36%

```python
curve = pd.DataFrame(rmsle_val)
curve.plot(figsize=(8,5));
```

```python
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor  
from sklearn.metrics import mean_squared_error, mean_squared_log_error
```

```python
knn = KNeighborsRegressor(n_neighbors = 20)

knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
knn.score(X_test, y_test)
```

Tasa de Precisión del modelo 25.6%

```python
from sklearn.metrics import r2_score, mean_absolute_error, explained_variance_score


r2_valid = r2_score(y_test, y_pred)
mae_valid = mean_absolute_error(y_test, y_pred)
evs_valid = explained_variance_score(y_test, y_pred, multioutput='uniform_average')
rmse_valid = np.sqrt(mean_squared_error(y_test, y_pred))
rmsle_valid = np.sqrt(mean_squared_log_error(y_test, y_pred))

print('R2 Valid:',r2_valid)
print('EVS Valid:', evs_valid)
print('MAE Valid:', mae_valid)
print('RMSE Valid:',rmse_valid)
print('RMSLE Valid:', rmsle_valid)
```
