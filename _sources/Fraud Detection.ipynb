## AnÃ¡lisis Bi Variado

```python
bivar = [('C1', 'isFraud'), ('C11','isFraud'),('C14','isFraud'),('V95','isFraud'),('V100','isFraud')]

for pair in bivar:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x=pair[0], y=pair[1], color='skyblue')
    plt.title(f'Scatterplot de {pair[0]} vs {pair[1]}')
    plt.xlabel(pair[0])
    plt.ylabel(pair[1])
    plt.show()
 
    plt.figure(figsize=(10, 6))
    sns.regplot(data=df, x=pair[0], y=pair[1], color= 'skyblue', line_kws={"color":"blue"})
    plt.title(f'Regplot de {pair[0]} vs {pair[1]}')
    plt.xlabel(pair[0])
    plt.ylabel(pair[1])
    plt.show()
```

Con el anÃ¡lisis bivariado podemos comprender cÃ³mo se relacionan las variables entre sÃ­ y la relaciÃ³n entre las variables independientes presentes en el conjunto de datos  y para identificar patrones de tendencia lineal.

La parte principal del grÃ¡fico es un conjunto de puntos dispersos en un plano cartesiano. Cada punto representa un par de valores de dos variables diferentes que estÃ¡s comparando.La lÃ­nea que atraviesa los puntos en el grÃ¡fico es la lÃ­nea de regresiÃ³n. Esta lÃ­nea se ajusta a los datos de manera que represente la mejor estimaciÃ³n de la relaciÃ³n entre las dos variables. En un grÃ¡fico de dispersiÃ³n, la lÃ­nea de regresiÃ³n se utiliza para modelar una posible tendencia lineal en los datos. La direcciÃ³n de la lÃ­nea de regresiÃ³n indica si hay una relaciÃ³n positiva o negativa entre las dos variables. Si la lÃ­nea sube hacia la derecha, hay una correlaciÃ³n positiva, lo que significa que cuando una variable aumenta, la otra tambiÃ©n tiende a aumentar. Si la lÃ­nea desciende hacia la derecha, hay una correlaciÃ³n negativa, lo que significa que cuando una variable aumenta, la otra tiende a disminuir.

## ImputaciÃ³n de datos faltantes con la media

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df_encoded_imputed = imputer.fit_transform(df_encoded)
df_imputed = pd.DataFrame(df_encoded_imputed, columns=df_encoded.columns)
```

```python
df_imputed.head()
```

Como el VIF no me terminÃ³ de cargar con todas la variables, se procede a realizar un anÃ¡lisis de correlaciÃ³n para eliminar las variables que estÃ©n mÃ¡s relacionadas. Se establece un umbral del 70%


```python
import pandas as pd


# Crear una copia del DataFrame original
df = df_imputed.copy()

# Obtener la matriz de correlaciÃ³n
correlation_matrix = df_imputed.corr()

# Obtener una lista de pares de columnas con alta correlaciÃ³n
high_correlation_pairs = []

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) >= 0.7:
            high_correlation_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j]))

# Crear una lista de columnas para eliminar
columns_to_remove = set()

for pair in high_correlation_pairs:
    columns_to_remove.add(pair[0])  # Agregar la primera columna del par
    columns_to_remove.add(pair[1])  # Agregar la segunda columna del par

# Eliminar las columnas de la lista 'columns_to_remove' del DataFrame
df.drop(columns=columns_to_remove, inplace=True)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# Crear un pipeline para cada modelo
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier(n_neighbors=5))
])

ridge = Pipeline([
    ('scaler', StandardScaler()),
    ('ridge', Ridge())
])

lasso = Pipeline([
    ('scaler', StandardScaler()),
    ('lasso', Lasso())
])

naiveBayes = Pipeline([
    ('naiveBayes', GaussianNB())
])

xgb = Pipeline([
    ('xgb', XGBClassifier())
])

svm = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC())
])

mlp = Pipeline([
    ('scaler', StandardScaler()),
    ('mlp', MLPClassifier())
])
```

Se definen los hiperparÃ¡metros para cada modelo:

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

# HiperparÃ¡metros para KNeighborsClassifier
knn_param_grid = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}

# HiperparÃ¡metros para Ridge
ridge_param_grid = {
    'ridge__alpha': [0.1, 1.0, 10.0]
}
# HiperparÃ¡metros para Lasso
lasso_param_grid = {
    'lasso__alpha': [0.1, 1.0, 10.0]
}

# HiperparÃ¡metros para XGBClassifier
xgb_param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5],
    'min_child_weight': [1, 2, 3],
    'subsample': [0.8, 0.9, 1.0]
}

# HiperparÃ¡metros para SVC
svc_param_grid = {
    'C': [0.1, 1.0, 10.0],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': [0.1, 1.0, 'auto']
}

# HiperparÃ¡metros para MLPClassifier
mlp_param_grid = {
    'hidden_layer_sizes': [(64, 32), (128, 64, 32), (256,)],
    'activation': ['relu', 'tanh', 'logistic'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate_init': [0.001, 0.01, 0.1]
}
```
